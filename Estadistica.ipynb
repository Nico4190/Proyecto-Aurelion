# Antigüedad de clientes (ventas.xlsx)
Analizamos la antigüedad de cada cliente a partir de su primera compra registrada en `df_ventas['fecha']`. Incluimos:
- Histograma de antigüedad (días) con línea de media
- Cohortes por rangos de antigüedad
- Top 10 clientes más antiguos y métricas resumen

```python
# Antigüedad de clientes usando df_ventas['fecha']
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Asegurar tipo datetime en la columna fecha
if 'fecha' not in df_ventas.columns:
    raise KeyError("La columna 'fecha' no existe en df_ventas. Verifica el archivo ventas.xlsx")

ventas = df_ventas.copy()
ventas['fecha'] = pd.to_datetime(ventas['fecha'], errors='coerce')
ventas = ventas.dropna(subset=['fecha'])

# 2) Consolidar por cliente: primera y última compra
agr = ventas.groupby('nombre_cliente', as_index=False).agg(
    primer_compra=('fecha', 'min'),
    ultima_compra=('fecha', 'max'),
    compras=('fecha', 'count')
)

# 3) Calcular antigüedad a una fecha fija (fin de 2024) y ventana de actividad observada
hoy = pd.Timestamp('2024-12-31').normalize() # Changed 'hoy' to end of 2024
agr['antiguedad_dias_fin_2024'] = (hoy - agr['primer_compra']).dt.days
agr['antiguedad_dias_actividad'] = (agr['ultima_compra'] - agr['primer_compra']).dt.days
agr['antiguedad_meses_fin_2024'] = (agr['antiguedad_dias_fin_2024'] / 30.44).round(1)

# 4) Cohortes de antigüedad
bins = [-1, 30, 90, 180, 365, 730, np.inf]
labels = ['0-30d', '31-90d', '91-180d', '181-365d', '1-2 años', '2+ años']
agr['cohorte_antiguedad'] = pd.cut(agr['antiguedad_dias_fin_2024'], bins=bins, labels=labels) # Use new column for cohorts

# 5) Resumen estadístico
prom = agr['antiguedad_dias_fin_2024'].mean() # Use new column for stats
med = agr['antiguedad_dias_fin_2024'].median() # Use new column for stats
mx = agr['antiguedad_dias_fin_2024'].max() # Use new column for stats
print("Clientes únicos:", len(agr))
print(f"Antigüedad media (fin 2024): {prom:.1f} días | mediana: {med:.0f} días | máx: {mx} días")
print("\nTop 10 clientes más antiguos (a fin de 2024):")
print(agr.sort_values('antiguedad_dias_fin_2024', ascending=False) # Use new column for sorting
        .loc[:, ['nombre_cliente', 'primer_compra', 'ultima_compra', 'antiguedad_dias_fin_2024', 'compras']] # Use new column in display
        .head(10))

# 6) Visualizaciones
fig, axes = plt.subplots(1, 2, figsize=(16,6))

# Histograma de antigüedad (días) con línea de media
sns.histplot(agr['antiguedad_dias_fin_2024'], bins=20, kde=False, color='#4C78A8', ax=axes[0]) # Use new column for plot
axes[0].axvline(prom, color='red', linestyle='--', linewidth=2, label=f"Media: {prom:.0f} días")
axes[0].set_title('Distribución de antigüedad de clientes (días a fin de 2024)')
axes[0].set_xlabel('Antigüedad en días desde 1ª compra (a fin de 2024)')
axes[0].set_ylabel('Frecuencia')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Barras por cohortes
coh = agr['cohorte_antiguedad'].value_counts().reindex(labels)
axes[1].bar(coh.index.astype(str), coh.values, color='#F58518', edgecolor='black')
axes[1].set_title('Cohortes por antigüedad (a fin de 2024)')
axes[1].set_xlabel('Rango de antigüedad')
axes[1].set_ylabel('Número de clientes')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# 7) Resultado listo en DataFrame `agr`
print("\nVista previa de la tabla de antigüedad (primeras 5 filas):")
print(agr.head())
```

# Relación entre fecha de alta de clientes y fecha de ventas
A continuación relacionamos la fecha de alta registrada en `clientes.xlsx` con las fechas de `ventas.xlsx` para medir:
- Días desde el alta hasta cada venta
- Días desde el alta hasta la primera compra por cliente
Incluye histogramas con línea de media y un resumen estadístico.

```python
# Relación entre fecha de alta (clientes) y fecha de ventas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1) Detectar claves y columnas de fecha ---
clientes = df_clientes.copy()
ventas = df_ventas.copy()

# Posibles claves para unir
claves_posibles = [
    ('nombre_cliente', 'nombre_cliente'),
    ('id_cliente', 'id_cliente'),
    ('cliente_id', 'cliente_id')
]
clave = None
for c_cli, c_ven in claves_posibles:
    if c_cli in clientes.columns and c_ven in ventas.columns:
        clave = (c_cli, c_ven)
        break
if clave is None:
    raise KeyError("No se encontró una clave común para unir df_clientes y df_ventas (intente 'nombre_cliente' o 'id_cliente').")

# Posibles nombres de 'fecha de alta' en clientes
candidatas_alta = [
    'fecha_alta', 'fecha_registro', 'alta', 'fecha_de_alta', 'fecha_ingreso',
    'fecha_alta_cliente'
]
col_alta = next((c for c in candidatas_alta if c in clientes.columns), None)
if col_alta is None:
    raise KeyError("No se encontró columna de fecha de alta en df_clientes (pruebe: 'fecha_alta', 'fecha_registro', 'alta', ...)")

# Columna de fecha de venta
col_venta = 'fecha'
if col_venta not in ventas.columns:
    raise KeyError("No existe columna 'fecha' en df_ventas.")

# --- 2) Normalizar fechas ---
clientes[col_alta] = pd.to_datetime(clientes[col_alta], errors='coerce')
ventas[col_venta] = pd.to_datetime(ventas[col_venta], errors='coerce')
clientes = clientes.dropna(subset=[col_alta])
ventas = ventas.dropna(subset=[col_venta])

# --- 3) Unir ventas con fecha de alta ---
df_mv = ventas.merge(clientes[[clave[0], col_alta]], left_on=clave[1], right_on=clave[0], how='inner', suffixes=('_venta','_cliente'))

# --- 4) Días desde el alta hasta cada venta ---
df_mv['dias_desde_alta'] = (df_mv[col_venta] - df_mv[col_alta]).dt.days
# Filtrar valores negativos (ventas antes del alta) si existieran
valid = df_mv['dias_desde_alta'].dropna()
valid = valid[valid >= 0]

# --- 5) Días hasta la primera compra por cliente ---
primera_venta = (df_mv
                 .groupby(clave[1], as_index=False)
                 .agg(fecha_primera_venta=(col_venta,'min'), fecha_alta=(col_alta,'first')))
primera_venta['dias_alta_a_primera'] = (primera_venta['fecha_primera_venta'] - primera_venta['fecha_alta']).dt.days
primera_venta = primera_venta.dropna(subset=['dias_alta_a_primera'])
primera_venta = primera_venta[primera_venta['dias_alta_a_primera'] >= 0]

# --- 6) Resúmenes ---
mean_all = valid.mean() if len(valid) else np.nan
median_all = valid.median() if len(valid) else np.nan
mean_first = primera_venta['dias_alta_a_primera'].mean() if len(primera_venta) else np.nan
median_first = primera_venta['dias_alta_a_primera'].median() if len(primera_venta) else np.nan

print("Ventas consideradas:", len(valid))
print(f"Días desde alta a CADA venta -> media: {mean_all:.1f} | mediana: {median_all:.0f}")
print(f"Días desde alta a PRIMERA venta -> media: {mean_first:.1f} | mediana: {median_first:.0f}")

# --- 7) Visualizaciones ---
fig, axes = plt.subplots(1, 2, figsize=(16,6))

# Histograma de días hasta cada venta
sns.histplot(valid, bins=30, color="#4C78A8", ax=axes[0])
if not np.isnan(mean_all):
    axes[0].axvline(mean_all, color='red', linestyle='--', linewidth=2, label=f"Media: {mean_all:.0f}d")
axes[0].set_title('Días desde alta hasta cada venta')
axes[0].set_xlabel('Días')
axes[0].set_ylabel('Frecuencia')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Histograma de días hasta la primera venta
sns.histplot(primera_venta['dias_alta_a_primera'], bins=30, color="#F58518", ax=axes[1])
if not np.isnan(mean_first):
    axes[1].axvline(mean_first, color='red', linestyle='--', linewidth=2, label=f"Media: {mean_first:.0f}d")
axes[1].set_title('Días desde alta hasta la PRIMERA venta')
axes[1].set_xlabel('Días')
axes[1].set_ylabel('Clientes')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Tabla: Top 10 más rápidos en comprar tras el alta
print("\nTop 10 clientes que compraron más rápido tras el alta:")
if len(primera_venta):
    cols_show = [clave[1], 'fecha_alta', 'fecha_primera_venta', 'dias_alta_a_primera']
    print(primera_venta.sort_values('dias_alta_a_primera').loc[:, cols_show].head(10))
else:
    print("No hay datos suficientes para calcular la primera compra.")
```

# Categorías más vendidas
Calculamos las categorías con mayor cantidad vendida a partir de `df_detalles_ventas`.
Si la columna de categoría no está en ese DataFrame, intentamos obtenerla desde `df_productos` mediante una unión por `nombre_producto`. Luego graficamos el Top 10 y mostramos porcentajes.

```python
# Top 10 categorías más vendidas (por cantidad)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Copias de trabajo
_det = df_detalles_ventas.copy()
_prod = 'df_productos' in globals() and isinstance(df_productos, pd.DataFrame)

# Asegurar columna cantidad
if 'cantidad' not in _det.columns:
    _det['cantidad'] = 1
_det['cantidad'] = pd.to_numeric(_det['cantidad'], errors='coerce').fillna(0)

# Intentar detectar columna de categoría en detalle_ventas
candidatas_cat = [
    'categoria', 'categoría', 'categoria_producto', 'categoria_nombre',
    'category', 'Categoria', 'Categoría', 'nombre_categoria'
]
col_cat_det = next((c for c in candidatas_cat if c in _det.columns), None)

# Si no existe en detalles, intentar obtener desde productos via merge por nombre_producto
if col_cat_det is None:
    if _prod:
        # Detectar columna de categoría en productos
        col_cat_prod = next((c for c in candidatas_cat if c in df_productos.columns), None)
        if col_cat_prod is None:
            raise KeyError("No se encontró una columna de categoría en detalle_ventas ni en productos.")
        # Necesitamos clave de unión
        if 'nombre_producto' in _det.columns and 'nombre_producto' in df_productos.columns:
            tmp = _det[['nombre_producto', 'cantidad']].merge(
                df_productos[['nombre_producto', col_cat_prod]],
                on='nombre_producto', how='left'
            )
            tmp = tmp.rename(columns={col_cat_prod: 'categoria_final'})
        else:
            raise KeyError("No se puede unir detalle_ventas y productos: falta 'nombre_producto' como clave en alguno de los DataFrames.")
    else:
        raise KeyError("No se encontró columna de categoría en detalle_ventas y no está disponible df_productos para unir.")
else:
    tmp = _det.rename(columns={col_cat_det: 'categoria_final'})[['categoria_final', 'cantidad']]

# Limpiar categorías vacías
tmp['categoria_final'] = tmp['categoria_final'].astype('string').fillna('Sin categoría')

# Agregar por categoría
cat_agg = tmp.groupby('categoria_final', as_index=False)['cantidad'].sum().sort_values('cantidad', ascending=False)

# Guardar resultado para reutilizar
try:
    df_categorias_ventas = cat_agg.copy()
except Exception:
    pass

# Top 10
top_n = 6
top_cat = cat_agg.head(top_n)

plt.figure(figsize=(12, 7))
sns.barplot(x='cantidad', y='categoria_final', data=top_cat, orient='h', color='#4C78A8', edgecolor='black')
plt.title(f'Top {top_n} Categorías Más Vendidas (por unidades)', fontsize=15, fontweight='bold')
plt.xlabel('Unidades vendidas')
plt.ylabel('Categoría')

# Anotar valores
for i, v in enumerate(top_cat['cantidad'].values):
    plt.text(v + max(top_cat['cantidad']) * 0.01, i, f"{int(v)}", va='center', fontweight='bold')

plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

# Tabla con porcentajes
total = cat_agg['cantidad'].sum()
tabla = top_cat.assign(porcentaje=(top_cat['cantidad'] / total) * 100)
print("TOP Categorías más vendidas (unidades y % del total):")
print(tabla.rename(columns={'categoria_final': 'Categoría', 'cantidad': 'Unidades', 'porcentaje': '%'}))
print(f"\nTotal unidades vendidas: {total}")
```

```python
# 1. Asegurar que el dataframe df_merged contenga las columnas 'categoria', 'importe' y 'cantidad'.
# Si no existe, fusionar df_detalles_ventas con df_productos usando 'nombre_producto' como clave para obtener la columna 'categoria'.
if 'categoria' not in df_detalles_ventas.columns:
    if 'df_productos' in globals() and isinstance(df_productos, pd.DataFrame):
        df_merged = pd.merge(df_detalles_ventas, df_productos[['nombre_producto', 'categoria']], on='nombre_producto', how='left')
    else:
        raise ValueError("df_productos no está disponible para fusionar y obtener la categoría.")
else:
    df_merged = df_detalles_ventas.copy()

# Ensure 'importe' and 'cantidad' columns exist
if 'importe' not in df_merged.columns:
    raise ValueError("La columna 'importe' no existe en el dataframe de detalles de ventas.")
if 'cantidad' not in df_merged.columns:
    # Assuming 'cantidad' might be missing, try to use 1 as default if not present
    df_merged['cantidad'] = 1
    print("Warning: 'cantidad' column not found in df_detalles_ventas. Using 1 for each item in a sale.")

# 2. Agrupar el dataframe resultante por 'categoria' y sumar las columnas 'importe' y 'cantidad'.
categoria_performance = df_merged.groupby('categoria').agg(
    importe_total=('importe', 'sum'),
    cantidad_total=('cantidad', 'sum')
).reset_index()

# 3. Crear un gráfico de dispersión para visualizar la relación entre importe total y cantidad total por categoría.
plt.figure(figsize=(12, 7))
sns.scatterplot(x='cantidad_total', y='importe_total', hue='categoria', data=categoria_performance, s=100) # Use hue for color by category

# 4. Añadir etiquetas y título al gráfico.
plt.title('Rendimiento de Categorías: Importe Total vs. Cantidad Total Vendida', fontsize=16)
plt.xlabel('Cantidad Total Vendida', fontsize=12)
plt.ylabel('Importe Total Vendido', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# 5. Mostrar el gráfico con leyenda.
plt.legend(title='Categoría', bbox_to_anchor=(1.05, 1), loc='upper left') # Move legend outside
plt.tight_layout()
plt.show()

# Display the aggregated data
display(categoria_performance)
```

```python
from geopy.geocoders import Nominatim
import folium

# Initialize Nominatim geocoder
geolocator = Nominatim(user_agent="city_geocoder")

# Get unique cities from the dataframe
cities = df_top_categoria_ciudad['ciudad'].unique()

# Create a dictionary to store coordinates
city_coordinates = {}

# Get coordinates for each city
for city in cities:
    try:
        location = geolocator.geocode(city)
        if location:
            city_coordinates[city] = (location.latitude, location.longitude)
            print(f"Coordinates for {city}: {city_coordinates[city]}")
        else:
            print(f"Could not find coordinates for {city}")
    except Exception as e:
        print(f"Error getting coordinates for {city}: {e}")

# Merge the top categories with their coordinates
df_top_categoria_ciudad_geo = df_top_categoria_ciudad.copy()
df_top_categoria_ciudad_geo['latitude'] = df_top_categoria_ciudad_geo['ciudad'].apply(lambda x: city_coordinates.get(x, (None, None))[0])
df_top_categoria_ciudad_geo['longitude'] = df_top_categoria_ciudad_geo['ciudad'].apply(lambda x: city_coordinates.get(x, (None, None))[1])

# Drop rows where coordinates were not found
df_top_categoria_ciudad_geo.dropna(subset=['latitude', 'longitude'], inplace=True)

# Create a Folium map centered around the average coordinates
if not df_top_categoria_ciudad_geo.empty:
    center_lat = df_top_categoria_ciudad_geo['latitude'].mean()
    center_lon = df_top_categoria_ciudad_geo['longitude'].mean()
    m = folium.Map(location=[center_lat, center_lon], zoom_start=7)

    # Add markers for each city with the top category
    for index, row in df_top_categoria_ciudad_geo.iterrows():
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=f"<b>City:</b> {row['ciudad']}<br><b>Top Category:</b> {row['categoria']}<br><b>Quantity:</b> {row['cantidad']}",
            tooltip=row['ciudad']
        ).add_to(m)

    # Display the map
    display(m)
else:
    print("No cities with valid coordinates to display on the map.")
```

## Evolución de ventas en el tiempo

### Subtask:
Analizar la tendencia de las ventas a lo largo del tiempo (por ejemplo, mensual o semanal).


**Reasoning**:
Ensure the 'fecha' column is in datetime format, extract the month and year, merge with details to get importe, group by the period, sum the importe, and sort by period to prepare for plotting the sales evolution over time.



```python
# 1. Asegurar que la columna 'fecha' en el dataframe df_ventas esté en formato datetime.
df_ventas['fecha'] = pd.to_datetime(df_ventas['fecha'])

# 2. Extraer el mes y el año de la columna 'fecha'
df_ventas['año_mes'] = df_ventas['fecha'].dt.to_period('M')

# 3. Fusionar df_ventas con df_detalles_ventas usando 'id_venta'
df_ventas_importe = pd.merge(df_ventas[['id_venta', 'año_mes']], df_detalles_ventas[['id_venta', 'importe']], on='id_venta', how='inner')

# 4. Agrupar por el período ('año_mes') y sumar el 'importe'
ventas_por_tiempo = df_ventas_importe.groupby('año_mes')['importe'].sum().reset_index()

# 5. Ordenar los resultados por período
ventas_por_tiempo['año_mes'] = ventas_por_tiempo['año_mes'].astype(str) # Convert period to string for plotting
ventas_por_tiempo = ventas_por_tiempo.sort_values('año_mes')

# 6. Crear un gráfico de líneas para visualizar la evolución de las ventas
plt.figure(figsize=(12, 6))
plt.plot(ventas_por_tiempo['año_mes'], ventas_por_tiempo['importe'], marker='o', linestyle='-')

# 7. Añadir etiquetas y título apropiados
plt.title('Evolución de Ventas a lo largo del Tiempo')
plt.xlabel('Período (Año-Mes)')
plt.ylabel('Importe Total de Ventas')
plt.xticks(rotation=45)
plt.grid(True)

# 8. Mostrar el gráfico.
plt.tight_layout()
plt.show()
```

The task requires calculating the average sales amount for each payment method and visualizing the result. This involves several data manipulation steps: calculating total sales per transaction, merging with sales data to link payment methods, grouping by payment method, calculating the mean importe, sorting, and finally plotting the results. All these steps can be combined into a single code block for efficiency.



```python
# 1. Calculate the total importe for each unique id_venta from the df_detalles_ventas DataFrame
ventas_con_importe = df_detalles_ventas.groupby('id_venta')['importe'].sum().reset_index()

# 2. Merge the resulting DataFrame with the df_ventas DataFrame using the id_venta column
df_ventas_metodo_pago = pd.merge(ventas_con_importe, df_ventas[['id_venta', 'medio_pago']], on='id_venta', how='inner')

# 3. Group the merged DataFrame by the medio_pago column and calculate the mean of the total importe for each payment method
promedio_importe_por_metodo = df_ventas_metodo_pago.groupby('medio_pago')['importe'].mean().reset_index()

# 4. Sort the resulting DataFrame by the average importe in descending order
promedio_importe_por_metodo = promedio_importe_por_metodo.sort_values('importe', ascending=False)

# 5. Create a bar plot to visualize the average sales amount per payment method.
plt.figure(figsize=(10, 6))
sns.barplot(x='medio_pago', y='importe', data=promedio_importe_por_metodo, palette='viridis')

# 6. Add appropriate labels to the x and y axes and a title to the plot.
plt.title('Importe Promedio de Ventas por Método de Pago')
plt.xlabel('Método de Pago')
plt.ylabel('Importe Promedio de Venta')

# 7. Display the plot.
plt.tight_layout()
plt.show()

# Display the calculated average importe per payment method
display(promedio_importe_por_metodo)
```
