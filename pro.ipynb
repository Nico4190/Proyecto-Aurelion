# Antigüedad de clientes (ventas.xlsx)
Analizamos la antigüedad de cada cliente a partir de su primera compra registrada en `df_ventas['fecha']`. Incluimos:
- Histograma de antigüedad (días) con línea de media
- Cohortes por rangos de antigüedad
- Top 10 clientes más antiguos y métricas resumen

```python
# Antigüedad de clientes usando df_ventas['fecha']
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Asegurar tipo datetime en la columna fecha
if 'fecha' not in df_ventas.columns:
    raise KeyError("La columna 'fecha' no existe en df_ventas. Verifica el archivo ventas.xlsx")

ventas = df_ventas.copy()
ventas['fecha'] = pd.to_datetime(ventas['fecha'], errors='coerce')
ventas = ventas.dropna(subset=['fecha'])

# 2) Consolidar por cliente: primera y última compra
agr = ventas.groupby('nombre_cliente', as_index=False).agg(
    primer_compra=('fecha', 'min'),
    ultima_compra=('fecha', 'max'),
    compras=('fecha', 'count')
)

# 3) Calcular antigüedad a hoy y ventana de actividad observada
hoy = pd.Timestamp.today().normalize()
agr['antiguedad_dias_hoy'] = (hoy - agr['primer_compra']).dt.days
agr['antiguedad_dias_actividad'] = (agr['ultima_compra'] - agr['primer_compra']).dt.days
agr['antiguedad_meses_hoy'] = (agr['antiguedad_dias_hoy'] / 30.44).round(1)

# 4) Cohortes de antigüedad
bins = [-1, 30, 90, 180, 365, 730, np.inf]
labels = ['0-30d', '31-90d', '91-180d', '181-365d', '1-2 años', '2+ años']
agr['cohorte_antiguedad'] = pd.cut(agr['antiguedad_dias_hoy'], bins=bins, labels=labels)

# 5) Resumen estadístico
prom = agr['antiguedad_dias_hoy'].mean()
med = agr['antiguedad_dias_hoy'].median()
mx = agr['antiguedad_dias_hoy'].max()
print("Clientes únicos:", len(agr))
print(f"Antigüedad media: {prom:.1f} días | mediana: {med:.0f} días | máx: {mx} días")
print("\nTop 10 clientes más antiguos (a hoy):")
print(agr.sort_values('antiguedad_dias_hoy', ascending=False)
        .loc[:, ['nombre_cliente', 'primer_compra', 'ultima_compra', 'antiguedad_dias_hoy', 'compras']]
        .head(10))

# 6) Visualizaciones
fig, axes = plt.subplots(1, 2, figsize=(16,6))

# Histograma de antigüedad (días) con línea de media
sns.histplot(agr['antiguedad_dias_hoy'], bins=20, kde=False, color='#4C78A8', ax=axes[0])
axes[0].axvline(prom, color='red', linestyle='--', linewidth=2, label=f"Media: {prom:.0f} días")
axes[0].set_title('Distribución de antigüedad de clientes (días)')
axes[0].set_xlabel('Antigüedad en días desde 1ª compra')
axes[0].set_ylabel('Frecuencia')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Barras por cohortes
coh = agr['cohorte_antiguedad'].value_counts().reindex(labels)
axes[1].bar(coh.index.astype(str), coh.values, color='#F58518', edgecolor='black')
axes[1].set_title('Cohortes por antigüedad')
axes[1].set_xlabel('Rango de antigüedad')
axes[1].set_ylabel('Número de clientes')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# 7) Resultado listo en DataFrame `agr`
print("\nVista previa de la tabla de antigüedad (primeras 5 filas):")
print(agr.head())
```

# Relación entre fecha de alta de clientes y fecha de ventas
A continuación relacionamos la fecha de alta registrada en `clientes.xlsx` con las fechas de `ventas.xlsx` para medir:
- Días desde el alta hasta cada venta
- Días desde el alta hasta la primera compra por cliente
Incluye histogramas con línea de media y un resumen estadístico.

```python
# Relación entre fecha de alta y ventas (barras + línea ECDF)
# Calcula días hasta la PRIMERA compra por cliente y grafica:
# - Barras: histograma de conteos
# - Línea: ECDF acumulada (%) en eje Y derecho

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

clientes = df_clientes.copy()
ventas = df_ventas.copy()

# 1) Detectar clave simple para unir
pares_clave = [
    ('nombre_cliente', 'nombre_cliente'),
    ('id_cliente', 'id_cliente'),
    ('cliente_id', 'cliente_id'),
]
clave = next(((c_cli, c_ven) for c_cli, c_ven in pares_clave if c_cli in clientes.columns and c_ven in ventas.columns), None)
if clave is None:
    raise KeyError("No hay una clave común simple para unir (intente 'nombre_cliente' o 'id_cliente').")

# 2) Detectar columna de alta en clientes y fecha en ventas (nombres más comunes)
col_alta = next((c for c in ['fecha_alta', 'fecha_registro', 'alta'] if c in clientes.columns), None)
if col_alta is None:
    raise KeyError("No se encontró la columna de alta en clientes (pruebe 'fecha_alta', 'fecha_registro' o 'alta').")
if 'fecha' not in ventas.columns:
    raise KeyError("No existe columna 'fecha' en ventas.")

# 3) Normalizar fechas y filtrar válidos
clientes[col_alta] = pd.to_datetime(clientes[col_alta], errors='coerce')
ventas['fecha'] = pd.to_datetime(ventas['fecha'], errors='coerce')
clientes = clientes.dropna(subset=[col_alta])
ventas = ventas.dropna(subset=['fecha'])

# 4) Unir y calcular días desde alta por venta
mv = ventas.merge(clientes[[clave[0], col_alta]], left_on=clave[1], right_on=clave[0], how='inner')
mv['dias_desde_alta'] = (mv['fecha'] - mv[col_alta]).dt.days
mv = mv.dropna(subset=['dias_desde_alta'])
mv = mv[mv['dias_desde_alta'] >= 0]

# 5) Días hasta la PRIMERA compra por cliente
primera = mv.groupby(clave[1], as_index=True)['dias_desde_alta'].min()

# 6) Métricas simples
n_clientes = primera.size
mean_first = float(primera.mean()) if n_clientes else np.nan
median_first = float(primera.median()) if n_clientes else np.nan
print(f"Clientes con al menos una compra tras el alta: {n_clientes}")
print(f"Días hasta la PRIMERA compra -> media: {mean_first:.1f} | mediana: {median_first:.0f}")

# 7) Barras (conteos) + línea ECDF (%)
fig, ax = plt.subplots(figsize=(12, 5))

# Barras: histograma de conteos
ax.hist(primera, bins=30, color="#4C78A8", edgecolor='black', alpha=0.85)
ax.set_xlabel('Días')
ax.set_ylabel('Clientes')
ax.grid(axis='y', alpha=0.3)

# Línea: ECDF en porcentaje (eje derecho)
ax2 = ax.twinx()
vals = np.sort(primera.values)
ecdf = (np.arange(1, len(vals) + 1) / len(vals)) * 100 if len(vals) else np.array([])
if len(vals):
    line_ecdf, = ax2.plot(vals, ecdf, color='orange', linewidth=2, label='ECDF (%)')
ax2.set_ylabel('% acumulado')
ax2.set_ylim(0, 100)

# Línea de media (eje izquierdo)
if not np.isnan(mean_first):
    line_mean = ax.axvline(mean_first, color='red', linestyle='--', linewidth=1.8, label=f"Media: {mean_first:.0f}d")

# Título y leyenda combinada
ax.set_title('Días desde alta hasta la PRIMERA compra (Barras + ECDF)')
handles1, labels1 = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
if handles1 or handles2:
    ax.legend(handles1 + handles2, labels1 + labels2, loc='lower right')

plt.tight_layout()
plt.show()
```

# Categorías más vendidas
Calculamos las categorías con mayor cantidad vendida a partir de `df_detalles_ventas`.
Si la columna de categoría no está en ese DataFrame, intentamos obtenerla desde `df_productos` mediante una unión por `nombre_producto`. Luego graficamos el Top 10 y mostramos porcentajes.

```python
# Categorías más vendidas (barras + línea de % acumulado)
# Agrupa por categoría (si está en detalles) o une con productos por nombre_producto
# y grafica el Top-N de unidades con una línea de porcentaje acumulado (Pareto).

# Preparar datos
_det = df_detalles_ventas.copy()
if 'cantidad' not in _det.columns:
    _det['cantidad'] = 1
_det['cantidad'] = pd.to_numeric(_det['cantidad'], errors='coerce').fillna(0)

# Elegir columna de categoría de forma simple
if 'categoria' in _det.columns:
    tmp = _det[['categoria', 'cantidad']].rename(columns={'categoria': 'Categoria'})
elif 'nombre_producto' in _det.columns and 'categoria' in df_productos.columns:
    tmp = _det[['nombre_producto', 'cantidad']].merge(
        df_productos[['nombre_producto', 'categoria']], on='nombre_producto', how='left'
    ).rename(columns={'categoria': 'Categoria'})[['Categoria', 'cantidad']]
else:
    raise KeyError("No se encontró una columna de categoría. Agrega 'categoria' en detalles o en productos y 'nombre_producto' para unir.")

# Agregar y ordenar
agg = (tmp.assign(Categoria=tmp['Categoria'].astype('string').fillna('Sin categoría'))
          .groupby('Categoria', as_index=False)['cantidad']
          .sum()
          .sort_values('cantidad', ascending=False))

# Top-N, porcentajes y % acumulado
TOP_N = 6
plot_df = agg.head(TOP_N).copy()
if len(plot_df) > 0:
    total = agg['cantidad'].sum()
    plot_df['Porcentaje'] = (plot_df['cantidad'] / total) * 100
    plot_df['Acumulado_%'] = plot_df['Porcentaje'].cumsum()

# Gráfico: barras (unidades) + línea (acumulado %)
fig, ax1 = plt.subplots(figsize=(12, 6))
x = range(len(plot_df))

# Barras de unidades
bars = ax1.bar(x, plot_df['cantidad'], color='#4C78A8', edgecolor='black', alpha=0.9, label='Unidades')
ax1.set_ylabel('Unidades vendidas')
ax1.grid(axis='y', alpha=0.3)

# Anotar valores sobre barras
for i, v in enumerate(plot_df['cantidad'].values):
    ax1.text(i, v + (plot_df['cantidad'].max() * 0.01), int(v), ha='center', va='bottom', fontweight='bold')

# Línea de % acumulado en eje derecho
ax2 = ax1.twinx()
if 'Acumulado_%' in plot_df.columns:
    ax2.plot(x, plot_df['Acumulado_%'], color='orange', linewidth=2.0, marker='o', label='% acumulado')
    ax2.set_ylim(0, 100)
ax2.set_ylabel('% acumulado')

# Eje X con etiquetas de categorías
ax1.set_xticks(x)
ax1.set_xticklabels(plot_df['Categoria'], rotation=45, ha='right')

# Título y leyenda combinada
ax1.set_title(f'Top {TOP_N} Categorías por unidades (barras) y % acumulado (línea)')
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
if handles1 or handles2:
    ax1.legend(handles1 + handles2, labels1 + labels2, loc='upper right')

plt.tight_layout()
plt.show()

# Tabla simple con porcentajes
print('TOP Categorías (unidades y %):')
cols = ['Categoria', 'cantidad'] + ([ 'Porcentaje', 'Acumulado_%'] if 'Porcentaje' in plot_df.columns else [])
print(plot_df[cols].rename(columns={'cantidad': 'Unidades'}))
```

# Boxplot de importes de venta
Mostramos la distribución de importes por línea de detalle (cantidad × precio_unitario). 
Incluimos:
- Boxplot en escala lineal y en log1p (útil si hay cola larga)
- Resumen de percentiles y métricas clave
Si se detecta una columna de id de venta, también calculamos importe por venta (suma de sus líneas).

```python
# Boxplot simple de importes por línea (cantidad × precio_unitario)
import pandas as pd
import matplotlib.pyplot as plt

# Preparar datos de manera sencilla
_det = df_detalles_ventas.copy()
if 'cantidad' not in _det.columns:
    _det['cantidad'] = 1

# Calcular importe por línea y limpiar
_det['cantidad'] = pd.to_numeric(_det['cantidad'], errors='coerce').fillna(0).astype(float)
_det['precio_unitario'] = pd.to_numeric(_det['precio_unitario'], errors='coerce').fillna(0).astype(float)
_det['importe_linea'] = _det['cantidad'] * _det['precio_unitario']

datos = (
    _det['importe_linea']
    .replace([float('inf'), float('-inf')], pd.NA)
    .dropna()
)
datos = datos[datos >= 0]

# Graficar un único boxplot (vertical)
plt.figure(figsize=(7, 5))
plt.boxplot(datos, vert=True, patch_artist=True,
            boxprops=dict(facecolor='white', color='black'),
            medianprops=dict(color='orange', linewidth=1.5),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'),
            flierprops=dict(marker='o', color='black', alpha=0.7))
plt.title('Boxplot de Importes de Venta')
plt.ylabel('Importe (ARS)')
plt.xlabel('')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
