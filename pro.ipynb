```python
# Importación de bibliotecas necesarias
import pandas as pd

```

```python
df_productos = pd.read_excel('./productos.xlsx')
df_ventas = pd.read_excel('./ventas.xlsx')
df_clientes = pd.read_excel('./clientes.xlsx')
df_detalles_ventas = pd.read_excel('./detalle_ventas.xlsx')
```

```python
#Mostrar las primeras filas de cada DataFrame
print("Productos:")
print(df_productos.head())
print("Ventas:")
print(df_ventas.head()) 
print("Clientes:")
print(df_clientes.head())   
print("Detalles de Ventas:")
print(df_detalles_ventas.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Estadistica descriptiva


```python
# ============================================
# Estadísticos descriptivos básicos del salario
# ============================================

# Calculamos un resumen estadístico general
stats_sal = df_detalles_ventas['precio_unitario'].describe()

# Medidas de tendencia central
media = df_detalles_ventas['precio_unitario'].mean()
mediana = df_detalles_ventas['precio_unitario'].median()
moda = df_detalles_ventas['precio_unitario'].mode().iloc[0]

# Medidas de dispersión
varianza = df_detalles_ventas['precio_unitario'].var()
desv_std = df_detalles_ventas['precio_unitario'].std()
rango = df_detalles_ventas['precio_unitario'].max() - df_detalles_ventas['precio_unitario'].min()

# --------------------------------------------
# Impresión de resultados
# --------------------------------------------

print("============================================")
print(" ANÁLISIS DESCRIPTIVO DEL PRECIO UNITARIO ")
print("============================================\n")

print("Resumen estadístico general:")
print(stats_sal)
print("\n--------------------------------------------")
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda:.2f}")
print("--------------------------------------------")
print(f"Varianza: {varianza:.2f}")
print(f"Desviación estándar: {desv_std:.2f}")
print(f"Rango (máx - mín): {rango:.2f}")
print("============================================")
```

# Visualizaciones

```python
# Gráfico de barras para métodos de pago (variable categórica)
plt.figure(figsize=(10,6))
conteo_medios = df_ventas['medio_pago'].value_counts()
plt.bar(conteo_medios.index, conteo_medios.values, color='green', edgecolor='black')
plt.title('Frecuencia de Métodos de Pago')
plt.xlabel('Métodos de Pago')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.grid(alpha=0.4, axis='y')
plt.tight_layout()
plt.show()

# Mostrar estadísticas de los métodos de pago
print("Distribución de métodos de pago:")
print(conteo_medios)
print(f"\nMétodo de pago más común: {conteo_medios.index[0]}")
print(f"Porcentaje del método más común: {(conteo_medios.iloc[0]/len(df_ventas)*100):.1f}%")
```

# Análisis de Clientes por Frecuencia de Compras

```python
# Análisis de frecuencia de compras por cliente
plt.figure(figsize=(12,8))
conteo_clientes = df_ventas['nombre_cliente'].value_counts()

# Mostrar solo los primeros 15 clientes para mejor visualización
top_clientes = conteo_clientes.head(15)

plt.bar(range(len(top_clientes)), top_clientes.values, color='skyblue', edgecolor='black')
plt.title('Top 15 Clientes por Frecuencia de Compras')
plt.xlabel('Clientes')
plt.ylabel('Número de Compras')
plt.xticks(range(len(top_clientes)), top_clientes.index, rotation=45, ha='right')
plt.grid(alpha=0.4, axis='y')
plt.tight_layout()
plt.show()

# Estadísticas descriptivas de los clientes
print("=" * 50)
print("    ANÁLISIS DE FRECUENCIA DE COMPRAS POR CLIENTE")
print("=" * 50)
print(f"Total de clientes únicos: {df_ventas['nombre_cliente'].nunique()}")
print(f"Total de ventas: {len(df_ventas)}")
print(f"Promedio de compras por cliente: {len(df_ventas) / df_ventas['nombre_cliente'].nunique():.2f}")
print("\nTop 10 clientes más frecuentes:")
print(conteo_clientes.head(10))
print(f"\nCliente con más compras: {conteo_clientes.index[0]} ({conteo_clientes.iloc[0]} compras)")
print(f"Porcentaje de ventas del top cliente: {(conteo_clientes.iloc[0]/len(df_ventas)*100):.1f}%")

# Distribución de frecuencias
print(f"\nDistribución de compras:")
print(f"Clientes con 1 compra: {(conteo_clientes == 1).sum()}")
print(f"Clientes con 2+ compras: {(conteo_clientes > 1).sum()}")
print(f"Clientes con 3+ compras: {(conteo_clientes > 2).sum()}")
print(f"Clientes con 4+ compras: {(conteo_clientes > 3).sum()}")
print(f"Clientes con 5+ compras: {(conteo_clientes > 4).sum()}")
print("=" * 50)
```

# Productos mas vendidos 

```python
# Histograma simple: Productos más vendidos vs Cantidad REAL
plt.figure(figsize=(14,8))

# Obtener los top 15 productos más vendidos (sumando cantidades reales)
productos_cantidad = df_detalles_ventas.groupby('nombre_producto')['cantidad'].sum().sort_values(ascending=False)
top_productos = productos_cantidad.head(20)

# Crear histograma de barras
plt.bar(range(len(top_productos)), top_productos.values, 
        color='steelblue', edgecolor='black', alpha=0.8)

plt.title('Top 15 Productos Más Vendidos (Unidades Reales)', fontsize=16, fontweight='bold')
plt.xlabel('Productos', fontsize=12)
plt.ylabel('Unidades Vendidas', fontsize=12)

# Configurar etiquetas del eje X
plt.xticks(range(len(top_productos)), top_productos.index, 
          rotation=45, ha='right', fontsize=10)

# Agregar valores encima de cada barra
for i, valor in enumerate(top_productos.values):
    plt.text(i, valor + 0.5, str(valor), ha='center', va='bottom', fontweight='bold')

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Mostrar tabla de resultados
print("TOP 20 PRODUCTOS MÁS VENDIDOS (UNIDADES REALES):")
print("-" * 60)
for i, (producto, cantidad) in enumerate(top_productos.items(), 1):
    print(f"{i:2d}. {producto:<35} | {cantidad:4d} unidades")
print("-" * 60)
```

# Relacion entre las ventas y el importe

```python
# Análisis de Relación: Ventas vs Importe por Producto
plt.figure(figsize=(15,10))

# Calcular cantidad vendida y importe total por producto (considerando cantidad y precio)
ventas_por_producto = df_detalles_ventas.groupby('nombre_producto').agg({
    'cantidad': 'sum',          # Total de unidades vendidas
    'precio_unitario': 'first'  # Precio unitario (para calcular importe)
}).reset_index()

# Calcular importe total = cantidad * precio_unitario
ventas_por_producto['Importe_Total'] = ventas_por_producto['cantidad'] * ventas_por_producto['precio_unitario']

# Renombrar columnas para claridad
ventas_por_producto = ventas_por_producto[['nombre_producto', 'cantidad', 'Importe_Total']]
ventas_por_producto.columns = ['Producto', 'Unidades_Vendidas', 'Importe_Total']

# Ordenar por unidades vendidas (descendente) y tomar top 15
top_productos_completo = ventas_por_producto.sort_values('Unidades_Vendidas', ascending=False).head(20)

# Crear gráfico de barras doble
x = range(len(top_productos_completo))
width = 0.35  # Ancho de las barras

# Normalizar importes para mejor visualización (dividir por 1000 si es necesario)
importes_normalizados = top_productos_completo['Importe_Total'] / 1000

fig, ax1 = plt.subplots(figsize=(15,8))

# Barras de unidades vendidas
barras1 = ax1.bar([i - width/2 for i in x], top_productos_completo['Unidades_Vendidas'], 
                  width, label='Unidades Vendidas', color='skyblue', alpha=0.8)

# Configurar primer eje Y (unidades)
ax1.set_xlabel('Productos', fontsize=12)
ax1.set_ylabel('Unidades Vendidas', color='blue', fontsize=12)
ax1.tick_params(axis='y', labelcolor='blue')

# Crear segundo eje Y para importes
ax2 = ax1.twinx()
barras2 = ax2.bar([i + width/2 for i in x], importes_normalizados, 
                  width, label='Importe Total (x1000)', color='lightcoral', alpha=0.8)

ax2.set_ylabel('Importe Total (Miles)', color='red', fontsize=12)
ax2.tick_params(axis='y', labelcolor='red')

# Configurar etiquetas del eje X
ax1.set_xticks(x)
ax1.set_xticklabels(top_productos_completo['Producto'], rotation=45, ha='right')

# Título y leyenda
plt.title('Relación Ventas vs Importe por Producto (Top 20)', fontsize=16, fontweight='bold')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()

# Mostrar tabla detallada
print("=" * 80)
print("    ANÁLISIS COMPLETO: VENTAS vs IMPORTE POR PRODUCTO")
print("=" * 80)
print(f"{'Rank':<4} | {'Producto':<25}  | {'Unidades':<10} | {'Importe':<12} | {'$/Unidad':<10}")
print("-" * 80)

for i, row in enumerate(top_productos_completo.itertuples(), 1):
    precio_promedio = row.Importe_Total / row.Unidades_Vendidas
    print(f"{i:<4} | {row.Producto[:24]:<25} | {row.Unidades_Vendidas:<10} | ${row.Importe_Total:<11.2f} | ${precio_promedio:<9.2f}")

print("-" * 80)

# Análisis de eficiencia (importe por unidad)
top_productos_completo['Importe_por_Unidad'] = top_productos_completo['Importe_Total'] / top_productos_completo['Unidades_Vendidas']
mejor_margen = top_productos_completo.loc[top_productos_completo['Importe_por_Unidad'].idxmax()]
mas_vendido = top_productos_completo.iloc[0]

print(f"\n📊 INSIGHTS CLAVE:")
print(f"🏆 Producto más vendido: {mas_vendido['Producto']} ({mas_vendido['Unidades_Vendidas']} unidades)")
print(f"💰 Mayor importe total: ${top_productos_completo['Importe_Total'].max():,.2f}")
print(f"💎 Mejor margen por unidad: {mejor_margen['Producto']} (${mejor_margen['Importe_por_Unidad']:.2f}/unidad)")
print("=" * 80)
```
