# Relacion entre las ventas y el importe

```python
# Análisis de Relación: Ventas vs Importe por Producto
plt.figure(figsize=(15,10))

# Calcular cantidad vendida y importe total por producto (considerando cantidad y precio)
ventas_por_producto = df_detalles_ventas.groupby('nombre_producto').agg({
    'cantidad': 'sum',          # Total de unidades vendidas
    'precio_unitario': 'first'  # Precio unitario (para calcular importe)
}).reset_index()

# Calcular importe total = cantidad * precio_unitario
ventas_por_producto['Importe_Total'] = ventas_por_producto['cantidad'] * ventas_por_producto['precio_unitario']

# Renombrar columnas para claridad
ventas_por_producto = ventas_por_producto[['nombre_producto', 'cantidad', 'Importe_Total']]
ventas_por_producto.columns = ['Producto', 'Unidades_Vendidas', 'Importe_Total']

# Ordenar por unidades vendidas (descendente) y tomar top 15
top_productos_completo = ventas_por_producto.sort_values('Unidades_Vendidas', ascending=False).head(20)

# Crear gráfico de barras doble
x = range(len(top_productos_completo))
width = 0.35  # Ancho de las barras

# Normalizar importes para mejor visualización (dividir por 1000 si es necesario)
importes_normalizados = top_productos_completo['Importe_Total'] / 1000

fig, ax1 = plt.subplots(figsize=(15,8))

# Barras de unidades vendidas
barras1 = ax1.bar([i - width/2 for i in x], top_productos_completo['Unidades_Vendidas'], 
                  width, label='Unidades Vendidas', color='skyblue', alpha=0.8)

# Configurar primer eje Y (unidades)
ax1.set_xlabel('Productos', fontsize=12)
ax1.set_ylabel('Unidades Vendidas', color='blue', fontsize=12)
ax1.tick_params(axis='y', labelcolor='blue')

# Crear segundo eje Y para importes
ax2 = ax1.twinx()
barras2 = ax2.bar([i + width/2 for i in x], importes_normalizados, 
                  width, label='Importe Total (x1000)', color='lightcoral', alpha=0.8)

ax2.set_ylabel('Importe Total (Miles)', color='red', fontsize=12)
ax2.tick_params(axis='y', labelcolor='red')

# Configurar etiquetas del eje X
ax1.set_xticks(x)
ax1.set_xticklabels(top_productos_completo['Producto'], rotation=45, ha='right')

# Título y leyenda
plt.title('Relación Ventas vs Importe por Producto (Top 20)', fontsize=16, fontweight='bold')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()

# Mostrar tabla detallada
print("=" * 80)
print("    ANÁLISIS COMPLETO: VENTAS vs IMPORTE POR PRODUCTO")
print("=" * 80)
print(f"{'Rank':<4} | {'Producto':<25}  | {'Unidades':<10} | {'Importe':<12} | {'$/Unidad':<10}")
print("-" * 80)

for i, row in enumerate(top_productos_completo.itertuples(), 1):
    precio_promedio = row.Importe_Total / row.Unidades_Vendidas
    print(f"{i:<4} | {row.Producto[:24]:<25} | {row.Unidades_Vendidas:<10} | ${row.Importe_Total:<11.2f} | ${precio_promedio:<9.2f}")

print("-" * 80)

# Análisis de eficiencia (importe por unidad)
top_productos_completo['Importe_por_Unidad'] = top_productos_completo['Importe_Total'] / top_productos_completo['Unidades_Vendidas']
mejor_margen = top_productos_completo.loc[top_productos_completo['Importe_por_Unidad'].idxmax()]
mas_vendido = top_productos_completo.iloc[0]

print(f"\n📊 INSIGHTS CLAVE:")
print(f"🏆 Producto más vendido: {mas_vendido['Producto']} ({mas_vendido['Unidades_Vendidas']} unidades)")
print(f"💰 Mayor importe total: ${top_productos_completo['Importe_Total'].max():,.2f}")
print(f"💎 Mejor margen por unidad: {mejor_margen['Producto']} (${mejor_margen['Importe_por_Unidad']:.2f}/unidad)")
print("=" * 80)
```

# Antigüedad de clientes (ventas.xlsx)
Analizamos la antigüedad de cada cliente a partir de su primera compra registrada en `df_ventas['fecha']`. Incluimos:
- Histograma de antigüedad (días) con línea de media
- Cohortes por rangos de antigüedad
- Top 10 clientes más antiguos y métricas resumen

```python
# Antigüedad de clientes usando df_ventas['fecha']
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Asegurar tipo datetime en la columna fecha
if 'fecha' not in df_ventas.columns:
    raise KeyError("La columna 'fecha' no existe en df_ventas. Verifica el archivo ventas.xlsx")

ventas = df_ventas.copy()
ventas['fecha'] = pd.to_datetime(ventas['fecha'], errors='coerce')
ventas = ventas.dropna(subset=['fecha'])

# 2) Consolidar por cliente: primera y última compra
agr = ventas.groupby('nombre_cliente', as_index=False).agg(
    primer_compra=('fecha', 'min'),
    ultima_compra=('fecha', 'max'),
    compras=('fecha', 'count')
)

# 3) Calcular antigüedad a hoy y ventana de actividad observada
hoy = pd.Timestamp.today().normalize()
agr['antiguedad_dias_hoy'] = (hoy - agr['primer_compra']).dt.days
agr['antiguedad_dias_actividad'] = (agr['ultima_compra'] - agr['primer_compra']).dt.days
agr['antiguedad_meses_hoy'] = (agr['antiguedad_dias_hoy'] / 30.44).round(1)

# 4) Cohortes de antigüedad
bins = [-1, 30, 90, 180, 365, 730, np.inf]
labels = ['0-30d', '31-90d', '91-180d', '181-365d', '1-2 años', '2+ años']
agr['cohorte_antiguedad'] = pd.cut(agr['antiguedad_dias_hoy'], bins=bins, labels=labels)

# 5) Resumen estadístico
prom = agr['antiguedad_dias_hoy'].mean()
med = agr['antiguedad_dias_hoy'].median()
mx = agr['antiguedad_dias_hoy'].max()
print("Clientes únicos:", len(agr))
print(f"Antigüedad media: {prom:.1f} días | mediana: {med:.0f} días | máx: {mx} días")
print("\nTop 10 clientes más antiguos (a hoy):")
print(agr.sort_values('antiguedad_dias_hoy', ascending=False)
        .loc[:, ['nombre_cliente', 'primer_compra', 'ultima_compra', 'antiguedad_dias_hoy', 'compras']]
        .head(10))

# 6) Visualizaciones
fig, axes = plt.subplots(1, 2, figsize=(16,6))

# Histograma de antigüedad (días) con línea de media
sns.histplot(agr['antiguedad_dias_hoy'], bins=20, kde=False, color='#4C78A8', ax=axes[0])
axes[0].axvline(prom, color='red', linestyle='--', linewidth=2, label=f"Media: {prom:.0f} días")
axes[0].set_title('Distribución de antigüedad de clientes (días)')
axes[0].set_xlabel('Antigüedad en días desde 1ª compra')
axes[0].set_ylabel('Frecuencia')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Barras por cohortes
coh = agr['cohorte_antiguedad'].value_counts().reindex(labels)
axes[1].bar(coh.index.astype(str), coh.values, color='#F58518', edgecolor='black')
axes[1].set_title('Cohortes por antigüedad')
axes[1].set_xlabel('Rango de antigüedad')
axes[1].set_ylabel('Número de clientes')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# 7) Resultado listo en DataFrame `agr`
print("\nVista previa de la tabla de antigüedad (primeras 5 filas):")
print(agr.head())
```

# Relación entre fecha de alta de clientes y fecha de ventas
A continuación relacionamos la fecha de alta registrada en `clientes.xlsx` con las fechas de `ventas.xlsx` para medir:
- Días desde el alta hasta cada venta
- Días desde el alta hasta la primera compra por cliente
Incluye histogramas con línea de media y un resumen estadístico.

```python
# Relación entre fecha de alta (clientes) y fecha de ventas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1) Detectar claves y columnas de fecha ---
clientes = df_clientes.copy()
ventas = df_ventas.copy()

# Posibles claves para unir
claves_posibles = [
    ('nombre_cliente', 'nombre_cliente'),
    ('id_cliente', 'id_cliente'),
    ('cliente_id', 'cliente_id')
]
clave = None
for c_cli, c_ven in claves_posibles:
    if c_cli in clientes.columns and c_ven in ventas.columns:
        clave = (c_cli, c_ven)
        break
if clave is None:
    raise KeyError("No se encontró una clave común para unir df_clientes y df_ventas (intente 'nombre_cliente' o 'id_cliente').")

# Posibles nombres de 'fecha de alta' en clientes
candidatas_alta = [
    'fecha_alta', 'fecha_registro', 'alta', 'fecha_de_alta', 'fecha_ingreso',
    'fecha_alta_cliente'
]
col_alta = next((c for c in candidatas_alta if c in clientes.columns), None)
if col_alta is None:
    raise KeyError("No se encontró columna de fecha de alta en df_clientes (pruebe: 'fecha_alta', 'fecha_registro', 'alta', ...)")

# Columna de fecha de venta
col_venta = 'fecha'
if col_venta not in ventas.columns:
    raise KeyError("No existe columna 'fecha' en df_ventas.")

# --- 2) Normalizar fechas ---
clientes[col_alta] = pd.to_datetime(clientes[col_alta], errors='coerce')
ventas[col_venta] = pd.to_datetime(ventas[col_venta], errors='coerce')
clientes = clientes.dropna(subset=[col_alta])
ventas = ventas.dropna(subset=[col_venta])

# --- 3) Unir ventas con fecha de alta ---
df_mv = ventas.merge(clientes[[clave[0], col_alta]], left_on=clave[1], right_on=clave[0], how='inner', suffixes=('_venta','_cliente'))

# --- 4) Días desde el alta hasta cada venta ---
df_mv['dias_desde_alta'] = (df_mv[col_venta] - df_mv[col_alta]).dt.days
# Filtrar valores negativos (ventas antes del alta) si existieran
valid = df_mv['dias_desde_alta'].dropna()
valid = valid[valid >= 0]

# --- 5) Días hasta la primera compra por cliente ---
primera_venta = (df_mv
                 .groupby(clave[1], as_index=False)
                 .agg(fecha_primera_venta=(col_venta,'min'), fecha_alta=(col_alta,'first')))
primera_venta['dias_alta_a_primera'] = (primera_venta['fecha_primera_venta'] - primera_venta['fecha_alta']).dt.days
primera_venta = primera_venta.dropna(subset=['dias_alta_a_primera'])
primera_venta = primera_venta[primera_venta['dias_alta_a_primera'] >= 0]

# --- 6) Resúmenes ---
mean_all = valid.mean() if len(valid) else np.nan
median_all = valid.median() if len(valid) else np.nan
mean_first = primera_venta['dias_alta_a_primera'].mean() if len(primera_venta) else np.nan
median_first = primera_venta['dias_alta_a_primera'].median() if len(primera_venta) else np.nan

print("Ventas consideradas:", len(valid))
print(f"Días desde alta a CADA venta -> media: {mean_all:.1f} | mediana: {median_all:.0f}")
print(f"Días desde alta a PRIMERA venta -> media: {mean_first:.1f} | mediana: {median_first:.0f}")

# --- 7) Visualizaciones ---
fig, axes = plt.subplots(1, 2, figsize=(16,6))

# Histograma de días hasta cada venta
sns.histplot(valid, bins=30, color="#4C78A8", ax=axes[0])
if not np.isnan(mean_all):
    axes[0].axvline(mean_all, color='red', linestyle='--', linewidth=2, label=f"Media: {mean_all:.0f}d")
axes[0].set_title('Días desde alta hasta cada venta')
axes[0].set_xlabel('Días')
axes[0].set_ylabel('Frecuencia')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Histograma de días hasta la primera venta
sns.histplot(primera_venta['dias_alta_a_primera'], bins=30, color="#F58518", ax=axes[1])
if not np.isnan(mean_first):
    axes[1].axvline(mean_first, color='red', linestyle='--', linewidth=2, label=f"Media: {mean_first:.0f}d")
axes[1].set_title('Días desde alta hasta la PRIMERA venta')
axes[1].set_xlabel('Días')
axes[1].set_ylabel('Clientes')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Tabla: Top 10 más rápidos en comprar tras el alta
print("\nTop 10 clientes que compraron más rápido tras el alta:")
if len(primera_venta):
    cols_show = [clave[1], 'fecha_alta', 'fecha_primera_venta', 'dias_alta_a_primera']
    print(primera_venta.sort_values('dias_alta_a_primera').loc[:, cols_show].head(10))
else:
    print("No hay datos suficientes para calcular la primera compra.")
```

# Categorías más vendidas
Calculamos las categorías con mayor cantidad vendida a partir de `df_detalles_ventas`.
Si la columna de categoría no está en ese DataFrame, intentamos obtenerla desde `df_productos` mediante una unión por `nombre_producto`. Luego graficamos el Top 10 y mostramos porcentajes.

```python
# Top 10 categorías más vendidas (por cantidad)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Copias de trabajo
_det = df_detalles_ventas.copy()
_prod = 'df_productos' in globals() and isinstance(df_productos, pd.DataFrame)

# Asegurar columna cantidad
if 'cantidad' not in _det.columns:
    _det['cantidad'] = 1
_det['cantidad'] = pd.to_numeric(_det['cantidad'], errors='coerce').fillna(0)

# Intentar detectar columna de categoría en detalle_ventas
candidatas_cat = [
    'categoria', 'categoría', 'categoria_producto', 'categoria_nombre',
    'category', 'Categoria', 'Categoría', 'nombre_categoria'
]
col_cat_det = next((c for c in candidatas_cat if c in _det.columns), None)

# Si no existe en detalles, intentar obtener desde productos via merge por nombre_producto
if col_cat_det is None:
    if _prod:
        # Detectar columna de categoría en productos
        col_cat_prod = next((c for c in candidatas_cat if c in df_productos.columns), None)
        if col_cat_prod is None:
            raise KeyError("No se encontró una columna de categoría en detalle_ventas ni en productos.")
        # Necesitamos clave de unión
        if 'nombre_producto' in _det.columns and 'nombre_producto' in df_productos.columns:
            tmp = _det[['nombre_producto', 'cantidad']].merge(
                df_productos[['nombre_producto', col_cat_prod]],
                on='nombre_producto', how='left'
            )
            tmp = tmp.rename(columns={col_cat_prod: 'categoria_final'})
        else:
            raise KeyError("No se puede unir detalle_ventas y productos: falta 'nombre_producto' como clave en alguno de los DataFrames.")
    else:
        raise KeyError("No se encontró columna de categoría en detalle_ventas y no está disponible df_productos para unir.")
else:
    tmp = _det.rename(columns={col_cat_det: 'categoria_final'})[['categoria_final', 'cantidad']]

# Limpiar categorías vacías
tmp['categoria_final'] = tmp['categoria_final'].astype('string').fillna('Sin categoría')

# Agregar por categoría
cat_agg = tmp.groupby('categoria_final', as_index=False)['cantidad'].sum().sort_values('cantidad', ascending=False)

# Guardar resultado para reutilizar
try:
    df_categorias_ventas = cat_agg.copy()
except Exception:
    pass

# Top 10
top_n = 6
top_cat = cat_agg.head(top_n)

plt.figure(figsize=(12, 7))
sns.barplot(x='cantidad', y='categoria_final', data=top_cat, orient='h', color='#4C78A8', edgecolor='black')
plt.title(f'Top {top_n} Categorías Más Vendidas (por unidades)', fontsize=15, fontweight='bold')
plt.xlabel('Unidades vendidas')
plt.ylabel('Categoría')

# Anotar valores
for i, v in enumerate(top_cat['cantidad'].values):
    plt.text(v + max(top_cat['cantidad']) * 0.01, i, f"{int(v)}", va='center', fontweight='bold')

plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

# Tabla con porcentajes
total = cat_agg['cantidad'].sum()
tabla = top_cat.assign(porcentaje=(top_cat['cantidad'] / total) * 100)
print("TOP Categorías más vendidas (unidades y % del total):")
print(tabla.rename(columns={'categoria_final': 'Categoría', 'cantidad': 'Unidades', 'porcentaje': '%'}))
print(f"\nTotal unidades vendidas: {total}")   
```
